```python
import polars as pl
import pandas as pd
from psifr import fr
data_file = "2011_Polyn.csv"
```

```python
raw = pd.read_csv(data_file)
raw
```

```python
data = fr.merge_free_recall(raw, list_keys=["task"], recall_keys=["period"], study_keys=["category"])
data.head(24)
```

```python
df = pl.DataFrame(data)
df = df.with_columns(
    pl.col(["input", "output", "category", "period", "prior_list", "prior_input"]).cast(pl.Int64)
)
df.head(24)
```
